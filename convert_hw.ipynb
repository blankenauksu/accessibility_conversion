# Convert SWP homework

Step 1 of 2: read a Scientific Word `.tex` file from `905_homework/` and write accessible homework files.

When a key is detected (after `\pagebreak`), this writes **two** `.tex` files in `905_homework/`:
- `STEM_hw.tex` — questions only (for students)
- `STEM_key_hw.tex` — full document with answer key

After running this notebook, **edit the outputs in TeXShop** if needed, then run **`compile_hw.ipynb`** to build PDFs in `canvas_pdfs/`.

Re-run this notebook only when the SWP source changes — it overwrites the output files.

```python
import sys
from pathlib import Path

ROOT = Path(".").resolve()
PIPELINE = ROOT / "accessible_homework"
MATERIALS = ROOT / "905_homework"

if str(PIPELINE) not in sys.path:
    sys.path.insert(0, str(PIPELINE))

from convert_swp_hw import convert_swp_hw, output_stem, peek_hw_source
```

```python
SOURCE = "hw1fall25.tex"  # SWP file in 905_homework/

_info = peek_hw_source(SOURCE, search_dirs=[MATERIALS])
ASSIGNMENT_NUMBER = 3  # edit assignment number
YEAR = 27  # two-digit year (27 → 2027)
OUTPUT_STEM = output_stem(_info["source_stem"], YEAR, ASSIGNMENT_NUMBER)
OUTPUT_STEM_KEY = f"{OUTPUT_STEM}_key"
HAS_KEY = _info["has_key"]

print(f"Source file: {_info['source']}")
print(f"Detected:    Assignment {_info['assignment_number']}, year 20{_info['year']:02d}")
print(f"Using:       Assignment {ASSIGNMENT_NUMBER}, year 20{YEAR:02d}")
print(f"Has key:     {HAS_KEY}")
print(f"Student:     {OUTPUT_STEM}_hw.tex")
if HAS_KEY:
    print(f"With key:    {OUTPUT_STEM_KEY}_hw.tex")
else:
    print("With key:    (skipped — no key in source)")
```

```python
convert_swp_hw(
    SOURCE,
    assignment_number=ASSIGNMENT_NUMBER,
    year=YEAR,
    search_dirs=[MATERIALS],
)
```
