# Compile accessible slides

Step 2 of 2: typeset an edited `*_slides.tex` file to a tagged PDF for Canvas.

Run this **after** converting (`convert_slides.ipynb`) and editing the `.tex` file.

You can also typeset in **TeXShop** (⌘T) — same engine, same result.

```python
import subprocess
import sys
from pathlib import Path

ROOT = Path(".").resolve()
PIPELINE = ROOT / "accessible_slides"
MATERIALS = ROOT / "905_materials"

if str(PIPELINE) not in sys.path:
    sys.path.insert(0, str(PIPELINE))

from validate_accessible_tex import print_validation_report, validate_accessible_tex

STEM = "slides1fall25"  # input stem: compiles STEM_slides.tex → STEM_slides.pdf
```

```python
tex = MATERIALS / f"{STEM}_slides.tex"
print_validation_report(validate_accessible_tex(tex))
```

```python
subprocess.run(
    ["bash", str(PIPELINE / "compile_pdf.sh"), STEM, str(MATERIALS)],
    check=False,
)

pdf = MATERIALS / f"{STEM}_slides.pdf"
if pdf.exists():
    print(f"\nTagged PDF ready: {pdf}")
    print("Upload this file to Canvas.")
else:
    print("\nBuild failed. Try compile_pdf.sh in Terminal, or TeXShop with LuaLaTeX.")
```
