# Compile accessible homework

Step 2 of 2: typeset edited `*_hw.tex` files to tagged PDFs for Canvas.

PDFs are saved in **`canvas_pdfs/`** (not in `905_homework/`).

Produces two PDFs when a key is detected:
- `STEM_hw.pdf` — questions only (for students)
- `STEM_key_hw.pdf` — questions plus answer key

Run this **after** converting (`convert_hw.ipynb`) and editing the `.tex` files.

```python
import subprocess
import sys
from pathlib import Path

ROOT = Path(".").resolve()
PIPELINE = ROOT / "accessible_homework"
MATERIALS = ROOT / "905_homework"
PDF_DIR = ROOT / "canvas_pdfs"

if str(PIPELINE) not in sys.path:
    sys.path.insert(0, str(PIPELINE))

from convert_swp_hw import output_stem, peek_hw_source
from validate_accessible_hw import print_validation_report, validate_accessible_hw
```

```python
SOURCE = "hw1fall25.tex"  # same as convert_hw.ipynb

_info = peek_hw_source(SOURCE, search_dirs=[MATERIALS])
ASSIGNMENT_NUMBER = 3  # same as convert_hw.ipynb
YEAR = 27  # same as convert_hw.ipynb
STEM = output_stem(_info["source_stem"], YEAR, ASSIGNMENT_NUMBER)
STEM_KEY = f"{STEM}_key"
HAS_KEY = _info["has_key"]

print(f"PDF folder:  {PDF_DIR}")
print(f"Student PDF: {STEM}_hw.pdf")
if HAS_KEY:
    print(f"Key PDF:     {STEM_KEY}_hw.pdf")
else:
    print("Key PDF:     (skipped — no key in source)")
```

```python
for stem in ([STEM, STEM_KEY] if HAS_KEY else [STEM]):
    tex = MATERIALS / f"{stem}_hw.tex"
    print(f"\n--- {tex.name} ---")
    print_validation_report(validate_accessible_hw(tex))
```

```python
for stem in ([STEM, STEM_KEY] if HAS_KEY else [STEM]):
    subprocess.run(
        ["bash", str(PIPELINE / "compile_pdf.sh"), stem, str(MATERIALS), str(PDF_DIR)],
        check=False,
    )
    pdf = PDF_DIR / f"{stem}_hw.pdf"
    if pdf.exists():
        print(f"Ready: {pdf}")
    else:
        print(f"Build failed: {pdf}")
```
