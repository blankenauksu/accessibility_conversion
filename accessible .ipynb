# Accessible slides workflow

This project uses **two notebooks**:

1. **`convert_slides.ipynb`** — convert SWP `.tex` → accessible `*_slides.tex`
2. Edit the `.tex` file in TeXShop (tags, structure)
3. **`compile_slides.ipynb`** — typeset `*_slides.tex` → tagged PDF for Canvas

Tools live in `accessible_slides/`. Source and output files live in `905_materials/`.
